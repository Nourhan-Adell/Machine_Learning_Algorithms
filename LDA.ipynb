## Iris data set (matrix 150*4) ---> Labeled Data (R4 ---> R2)

```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

## Load iris dataset

```python
iris = load_iris()
x = iris.data       #all the data in the matrix
y = iris.target     #label
print('Data: \n', x)
```

# Step 1: Standardization

```python
standarized = StandardScaler().fit_transform(x)
print("Standarized data: \n", standarized)
```

# Step 2: Computing the dimension mean vector
# Step 3: Computing the scatter matrix
# Step 4: Compute Eigen values and Eigen values
# Step 5: Construct the projection matrix
# Step 6: Project into a new feature space

```python
#Dimentional reduction by LDA 
LDA = LinearDiscriminantAnalysis(n_components = 2)
X_LDA = LDA.fit_transform(standarized, y)
print("Data after LDA: \n",X_LDA)
```

# Visualize the data


```python
plt.figure(figsize = (10,5))
plt.xlabel('LDA 1')
plt.ylabel('LDA 2')
plt.scatter(X_LDA[:,0], X_LDA[:,1], c=y)
plt.show()
```
